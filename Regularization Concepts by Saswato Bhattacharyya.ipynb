we can see that the value of Cost function is huge and this is a typical example of overfitting. Overfitting occurs when the model perfectly fits the training data but does not fit the test data well.

This problem can be solved by regularization so that the model goes from giving high variance to low variance.

Thus regularization techniques are used to reduce overfitting and get a more generalised model.
